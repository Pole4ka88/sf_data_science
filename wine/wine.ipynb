```python
import pandas as pd

wines = pd.read_csv('wine.csv')
wine = wines.copy()
display(wine)
```

```python
wine_drop = wine.drop_duplicates(inplace = False)
display(wine_drop)
```

```python
wine_is_null = wine_drop.isnull().mean()*100
wine_drop_null = wine_is_null[wine_is_null > 0].sort_values(ascending=False)
display(wine_drop_null)

wine_hist = wine_drop_null.index
wine_drop[wine_hist].hist()
```

```python
wine_drop = wine_drop.drop(['region_2'], axis=1)
display(wine_drop)
```

```python
values = {
    'designation': 'unknown',
    'taster_twitter_handle': 'unknown',
    'taster_name': 'unknown',
    'region_1': 'unknown',
    'price': wine_drop['price'].median(),
    'country': 'unknown',
    'province': 'unknown',
    'variety': 'unknown'}
    
fill_wine = wine_drop.fillna(values)   
fill_wine.isnull().mean() 
    
    
    
    
```
